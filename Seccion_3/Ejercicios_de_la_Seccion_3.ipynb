## 2.&nbsp;Ejercicios

```python
"""
Ejercicio 1: Cree una sesión de Spark con nombre Cap2 y asegúrese de que emplea
todos los cores disponibles para ejecutar en su ambiente de trabajo.
"""
spark = SparkSession.builder.master("local[*]").appName("Cap2").getOrCreate()
sc = spark.sparkContext
```

```python
"""
Ejercicio 2: Cree dos RDD vacíos, uno de ellos no debe contener particiones y
el otro debe tener 5 particiones. Utilice vías diferentes para crear cada RDD.
"""
empty_rdd_1 = sc.emptyRDD
empty_rdd_2 = sc.parallelize([], 5)
empty_rdd_2.collect()
```

```python
"""
Ejercicio 3: Cree un RDD que contenga los números primos que hay entre 1 y 20.
"""
primes_rdd = sc.parallelize([x for x in range(2, 20) if all(x % y for y in range(2, x//2))])
primes_rdd.collect()
```

```python
"""
Ejercicio 4 : Cree un nuevo RDD a partir del RDD creado en el ejercicio
anterior el cuál solo contenga los números primos mayores a 10.
"""
primes_gt_10_rdd = primes_rdd.filter(lambda x: x > 10)
primes_gt_10_rdd.collect()
```

```python
"""
Ejercicio 5.a: Cree un RDD a partir de este archivo de texto: el_valor_del_big_data.txt
en donde todo el documento esté contenido en un solo registro.
¿Cómo podría saber la dirección donde está guardado el archivo de texto
a partir del RDD creado?
"""
text_file_rdd_a = sc.wholeTextFiles('./el_valor_del_big_data.txt')
text_file_rdd_a.collect()
filepath = text_file_rdd_a.collect()[0][0]
filepath
```

```python
"""
Ejercicio 5.b: Si necesitara crear un RDD a partir del archivo de texto cargado
previamente en donde cada línea del archivo fuera un registro del RDD,
¿cómo lo haría?
"""
text_file_rdd_b = sc.textFile('./el_valor_del_big_data.txt')
text_file_rdd_b.collect()
```
